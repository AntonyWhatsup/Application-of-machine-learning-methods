```python
# Import podstawowych bibliotek
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ustawienia stylów wykresów
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (10, 5)
```

```python
# Wczytanie danych z pominięciem pierwszych 3 wierszy (metadane)
file_path = 'Data_Professional_Salary_Survey_Responses.xlsx'
df = pd.read_excel(file_path, skiprows=3)

# Ujednolicenie nazw kolumn (usunięcie spacji)
df.columns = df.columns.str.strip().str.replace(' ', '_')

# Wyświetlenie wymiarów danych
print("Kształt danych:", df.shape)
df.head(3)
```

```python
# Utworzenie kopii roboczej
df_clean = df.copy()

# Usunięcie wierszy, gdzie brakuje obu zmiennych docelowych
df_clean = df_clean.dropna(subset=['SalaryUSD', 'CareerPlansThisYear'], how='all')
```

```python
# Oczyszczanie SalaryUSD z wartości odstających (1% i 99% percentyl)
q_low = df_clean['SalaryUSD'].quantile(0.01)
q_hi  = df_clean['SalaryUSD'].quantile(0.99)

df_clean = df_clean[(df_clean['SalaryUSD'] > q_low) & (df_clean['SalaryUSD'] < q_hi)]

print("Zakres wynagrodzeń po czyszczeniu:", df_clean['SalaryUSD'].min(), "-", df_clean['SalaryUSD'].max())
```

```python
# Wypełnienie braków i konwersja na tekst
df_clean['CareerPlansThisYear'] = df_clean['CareerPlansThisYear'].fillna('Unknown').astype(str)

print("Najczęstsze plany zawodowe:\n", df_clean['CareerPlansThisYear'].value_counts().head())
```

```python
# Wizualizacja zmiennych docelowych
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# 1. Rozkład SalaryUSD
sns.histplot(df_clean['SalaryUSD'], bins=40, kde=True, ax=axes[0])
axes[0].set_title('Rozkład wynagrodzeń (SalaryUSD)')

# 2. Rozkład CareerPlansThisYear (Top 5 odpowiedzi)
top_plans = df_clean['CareerPlansThisYear'].value_counts().nlargest(5).index
sns.countplot(y='CareerPlansThisYear', data=df_clean[df_clean['CareerPlansThisYear'].isin(top_plans)], 
              ax=axes[1], order=top_plans)
axes[1].set_title('Top 5 planów zawodowych')

plt.tight_layout()
plt.show()
```
